_Початкова умова:_

Description
This is your first potion class in Hogwarts and professor gave you a homework to figure out what color potion will turn into if he'll mix it with some other potion. All potions have some color that written down as RGB color from [0, 0, 0] to [255, 255, 255]. To make task more complicated teacher will do few mixing and after will ask you for final color. Besides color you also need to figure out what volume will have potion after final mix.

Task
Based on your programming background you managed to figure that after mixing two potions colors will mix as if mix two RGB colors. For example, if you'll mix potion that have color [255, 255, 0] and volume 10 with one that have color [0, 254, 0] and volume 5, you'll get new potion with color [170, 255, 0] and volume 15. So you decided to create a class Potion that will have two properties: color (a list (a tuple in Python) with 3 integers) and volume (a number), and one method mix that will accept another Potion and return a mixed Potion.

Example
felix_felicis       =  Potion([255, 255, 255],  7)
shrinking_solution  =  Potion([ 51, 102,  51], 12)
new_potion          =  felix_felicis.mix(shrinking_solution)

new_potion.color   ==  [127, 159, 127]
new_potion.volume  ==  19
Note: Use ceiling when calculating the resulting potion's color.



_Код від ШІ:_

```python
import math

class Potion:
    def __init__(self, color, volume):
        self.color = color
        self.volume = volume

    def mix(self, other):
        new_color = [
            math.ceil((self.color[0] * self.volume + other.color[0] * other.volume) / (self.volume + other.volume)),
            math.ceil((self.color[1] * self.volume + other.color[1] * other.volume) / (self.volume + other.volume)),
            math.ceil((self.color[2] * self.volume + other.color[2] * other.volume) / (self.volume + other.volume))
        ]
        new_volume = self.volume + other.volume
        return Potion(new_color, new_volume)

# Приклад використання:
felix_felicis = Potion([255, 255, 255], 7)
shrinking_solution = Potion([51, 102, 51], 12)
new_potion = felix_felicis.mix(shrinking_solution)

print(new_potion.color)  # [127, 159, 127]
print(new_potion.volume)  # 19
```

_Початкова умова:_

Here's another staple for the functional programmer. You have a sequence of values and some predicate for those values. You want to get the longest prefix of elements such that the predicate is true for each element. We'll call this the takeWhile function. It accepts two arguments. The first is the sequence of values, and the second is the predicate function. The function does not change the value of the original sequence.

Example:
sequence : [2,4,6,8,1,2,5,4,3,2]
predicate: is an even number
result   : [2,4,6,8]
Your task is to implement the takeWhile function.

If you've got a span function lying around, this is a one-liner! Also, if you liked this problem, you'll surely love the dropWhile function.

_Код від ШІ:_

```python
def take_while(sequence, predicate):
    result = []
    for item in sequence:
        if predicate(item):
            result.append(item)
        else:
            break
    return result

# Example usage:
sequence = [2, 4, 6, 8, 1, 2, 5, 4, 3, 2]
predicate = lambda x: x % 2 == 0  # Predicate function for even numbers
result = take_while(sequence, predicate)
print(result)  # Output: [2, 4, 6, 8]
```

_Скорочена умова:_

You have a sequence of values and some predicate for those values. You want to get the longest prefix of elements such that the predicate is true for each element. We'll call this the takeWhile function. It accepts two arguments. The first is the sequence of values, and the second is the predicate function. The function does not change the value of the original sequence.

```python
def takeWhile(sequence, predicate):
    result = []
    for value in sequence:
        if predicate(value):
            result.append(value)
        else:
            break
    return result

# Example usage:
# Define a predicate function
def is_positive(x):
    return x > 0

# Test the takeWhile function
sequence = [1, 2, 3, 4, -1, -2, -3, 5, 6]
result = takeWhile(sequence, is_positive)
print(result)  # Output will be [1, 2, 3, 4]
```

Зі скороченою умовою ШІ також справився, проте створив функції, які в кінцевому результаті не потрібні були, тому код міг виглядати коротше.

_Відшліфований код:_

```python
def takeWhile(sequence, predicate):
    result = []
    for value in sequence:
        if predicate(value):
            result.append(value)
        else:
            break
    return result
```
